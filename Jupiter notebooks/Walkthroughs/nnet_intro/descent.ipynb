```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d


# Define the function to optimize
def f(x, y):
    return x ** 2 + y


# Define the gradient of the function
def grad_f(x, y):
    return [1.5 * x, 1.5 * y]

# Set the starting point
x = -10
y = 0

# Set the learning rate
lr = 0.01

# Set the number of iterations
n_iter = 100

# Initialize a list to store the values of x and y at each iteration
xs = []
ys = []
zs = []

# Perform gradient descent
for i in range(n_iter):
    # Compute the gradient
    grad = grad_f(x, y)

    # Update the variables
    x -= lr * grad[0]
    y -= lr * grad[1]

    # Store the values of x and y
    xs.append(x)
    ys.append(y)
    zs.append(f(x, y))

# Create a figure and a 3D axis
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(60, 50)

# Plot the function
X, Y = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))
Z = f(X, Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm', zorder=1)

# Plot the path of the optimization algorithm
ax.plot3D(xs, ys, zs, 'green', linewidth=3, zorder=10)

# Show the plot
plt.show()
```
